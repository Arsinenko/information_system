```python
import random
from faker import Faker 
import requests

def create_students():
    faker = Faker("ru_RU")

    for i in range(40):
        first_name = faker.first_name_male()
        middle_name = faker.middle_name_male()
        last_name = faker.last_name_male()
        group_id = random.randint(3, 10)
    # send request
        data = {
        "first_name": first_name,
        "middle_name": middle_name,
        "last_name": last_name,
        "group_id" : group_id
    }
        header = {
        "Content-Type": "application/json",
    }
        response = requests.post(json=data, url="http://localhost:8000/api/v1/create_student", headers=header)
        print(response.status_code)

create_students()
```

```python
def create_groups():
    names = ["Group A", "Group B", "Group C", "Group D", "Group E", "Group F", "Group G", "Group H", "Group I", "Group J"]
    for elem in names:
        data = {
            "group_name": elem
        }
        header = {
            "Content-Type": "application/json",
        }
        response = requests.post(json=data, url="http://localhost:8000/api/v1/create_group", headers=header)
        print(response.status_code)

create_groups()
```

```python
def create_teachers():
    faker = Faker("ru_RU")
    for i in range(10):
        first_name = faker.first_name_male()
        middle_name = faker.middle_name_male()
        last_name = faker.last_name_male()
        data = {
            "first_name": first_name,
            "middle_name": middle_name,
            "last_name": last_name
        }
        header = {
            "Content-Type": "application/json",
        }
        response = requests.post(json=data, url="http://localhost:8000/api/v1/create_teacher", headers=header)
        print(response.status_code)

create_teachers()
```

```python
def create_subjects():
    names = ["Math", "Physics", "Chemistry", "Biology", "History", "Geography", "English", "French", "German", "Spanish"]
    for elem in names:
        id_teacher = random.randint(1, 10)
        data = {
            "id_teacher": id_teacher,
            "subject_name": elem,
            "hours": random.randint(30, 100)
        }
        header = {
            "Content-Type": "application/json",
        }
        response = requests.post(json=data, url="http://localhost:8000/api/v1/create_subject", headers=header)
        print(response.status_code)

create_subjects()
```

```python
from faker import Faker
def push_attendance_records():
    faker = Faker("ru_RU")
    for i in range(100):
        id_subject = random.randint(1, 10)
        id_student = random.randint(1, 40)
        status = random.choice(["present", "absent"])
        pair_number = random.randint(1, 4)
        date = faker.date_between(start_date="-30d", end_date="today")

        data = {
        "subject_id": id_subject,
        "student_id": id_student, 
        "date": str(date),
        "pair_number": pair_number,
        "status": status
    }
        header = {
        "Content-Type": "application/json"
    }
        response = requests.post(json=data, url="http://localhost:8000/api/v1/create_attendance_record", headers=header)
        print(response.status_code)

push_attendance_records()
```
